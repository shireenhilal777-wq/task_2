# **Оптимизация типов pandas**
**1. Считайте датасет из файла Housing.csv (это данные о ценах домов) с помощью pandas**

```python
hs = pd.read_csv('/content/Housing.csv')
hs.head()
```

**2. Для каждого столбца определите оптимальный с точки зрения потребления памяти тип данных - напишите свои выводы в комментариях**

```python
# Проверка памяти перед оптимизацией
mem_before = hs.memory_usage(deep=True).sum()
print("Memory before optimization:",mem_before,"bytes")
```

```python
# показать текущие типы данных
print(hs.dtypes)
print(hs.describe(include = "all"))
```

**Вывод:**
- Числовые столбцы можно оптимизировать с помощью downcast до более компактных типов.
- Строковые столбцы с повторяющимися значениями можно перевести в category.
- Это уменьшит потребление памяти и сделает работу с датафреймом более эффективной.

**3. Поменяйте типы данных столбцов датафрейма на выбранные вами в прошлом пункте и сравните потребление памяти до и после оптимизации**

```python
# создать оптимизированную копию
hs_opt = hs.copy()

# numeric columns
hs_opt["price"] = hs_opt["price"].astype("int32")
hs_opt["area"] = hs_opt["area"].astype("int16")
hs_opt["bedrooms"] = hs_opt["bedrooms"].astype("int8")
hs_opt["bathrooms"] = hs_opt["bathrooms"].astype("int8")
hs_opt["stories"] = hs_opt["stories"].astype("int8")
hs_opt["parking"] = hs_opt["parking"].astype("int8")

# categorical columns
hs_opt["mainroad"] = hs_opt["mainroad"].astype("category")
hs_opt["guestroom"] = hs_opt["guestroom"].astype("category")
hs_opt["basement"] = hs_opt["basement"].astype("category")
hs_opt["hotwaterheating"] = hs_opt["hotwaterheating"].astype("category")
hs_opt["airconditioning"] = hs_opt["airconditioning"].astype("category")
hs_opt["prefarea"] = hs_opt["prefarea"].astype("category")
hs_opt["furnishingstatus"] = hs_opt["furnishingstatus"].astype("category")

# memory after optimization
mem_after = hs_opt.memory_usage(deep=True).sum()
print("Память после оптимизации:", mem_after, "байт")
print("Экономия памяти:", mem_before - mem_after, "байт")
print("Снижение:", round((mem_before - mem_after) / mem_before * 100, 2), "%")

# new types
print("\nНовые типы данных:")
print(hs_opt.dtypes)
```

**Вывод:**
 - Для столбцов price, area, bedrooms, bathrooms, stories и parking
 были выбраны более компактные целочисленные типы, так как диапазон их значений это позволяет.
- Столбцы mainroad, guestroom, basement, hotwaterheating, airconditioning, prefarea и furnishingstatus переведены в тип category, поскольку содержат
 небольшое число повторяющихся строковых значений.
- После изменения типов данных потребление памяти датафреймом уменьшилось.
