**4. Выведите самое популярное мужское и самое популярное женское имя на корабле**

```python
table["MostPopName"] = (
    table["Name"].str.split("(").str[1].str.split(")").str[0].fillna(
        table["Name"].str.split(",").str[1].str.split(".").str[1].str.strip()
    )
    .str.split().str[0]
)

print(table[table["Sex"] == "male"]["MostPopName"].value_counts().head(1))
print(table[table["Sex"] == "female"]["MostPopName"].value_counts().head(1))
```

**5. Выведите самое популярное мужское и самое популярное женское имя на корабле в каждом классе**

```python
popular_names = table.groupby(["Pclass", "Sex"])["MostPopName"].agg(lambda x: x.value_counts().idxmax())
print(popular_names)
```
